````python
import pandas as pd

# URL de la tabla
#url = "Ponme aquiiiii el liiiink"
url = "https://en.wikipedia.org/wiki/List_of_people_who_have_been_pied"

# Leer todas las tablas de la p치gina
tablas = pd.read_html(url)

# Seleccionamos la primera tabla (la principal del art칤culo)
df = tablas[0]

# Renombramos columnas para que sea m치s c칩modo trabajar
df.columns = ['Victim', 'Notoriety', 'Date', 'Details', 'Location', 'Perpetrator', 'Source']

# Mostrar las primeras filas del DataFrame
df.head()
````

````python
# Filtrar fechas con formato completo: "Month Day, Year" (ej. January 18, 1999)
fecha_completa_regex = r'^[A-Za-z]+ \d{1,2}, \d{4}$'
df_fechas = df[df['Date'].str.match(fecha_completa_regex, na=False)].copy()
# Convertir a datetime
df_fechas['Date'] = pd.to_datetime(df_fechas['Date'])
````

````python
# Obtener el mes m치s com칰n
mes_mas_comun = df_fechas['Date'].dt.month_name().value_counts()
mes_mas_comun
````

````python
anio_mas_comun = df_fechas['Date'].dt.year.value_counts()
anio_mas_comun
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

df_fechas['Year'] = df_fechas['Date'].dt.year
df_fechas['Month'] = df_fechas['Date'].dt.month_name()

# 游늰 Gr치fico de a침os con m치s pastelazos
plt.figure(figsize=(10, 5))
sns.countplot(data=df_fechas, x='Year', order=sorted(df_fechas['Year'].value_counts().index))
plt.title("游늰 A침os con m치s pastelazos (solo fechas completas)")
plt.xlabel("A침o")
plt.ylabel("N칰mero de casos")
plt.xticks(rotation=45)
plt.tight_layout()
# Guardar como imagen PNG
# plt.savefig("grafica_meses.png", dpi=300)
plt.show()
````

````python
# 游늱 Gr치fico de meses m치s comunes
plt.figure(figsize=(10, 5))
# Ordenar meses en orden calendario
meses_orden = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
sns.countplot(data=df_fechas, x='Month', order=meses_orden)
plt.title("游늱 Meses con m치s pastelazos (solo fechas completas)")
plt.xlabel("Mes")
plt.ylabel("N칰mero de casos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
# Filtrar solo los perpetradores frecuentes
filtro_perpetradores = ['Les Entartistes', 'A.J. Burnett', 'Aron Kay', 'Biotic Baking Brigade', 'No칢l Godin']
df_filtrado = df_fechas[df_fechas['Perpetrator'].isin(filtro_perpetradores)]

````
